```python
import requests
import pandas as pd
import time

def geocode_address(address, api_key):
    # Base URL for the OpenCage API
    url = 'https://api.opencagedata.com/geocode/v1/json'

    # Parameters for the API request
    params = {
        'q': address,       # The address to geocode
        'key': api_key,     # Your API key
        'limit': 1,         # Limit the number of results
        'no_annotations': 1 # Do not include additional annotations
    }

    # Make the API request
    response = requests.get(url, params=params)

    # Check if the request was successful
    if response.status_code == 200:
        data = response.json()

        if data['results']:
            # Extract the first result's geometry (latitude and longitude)
            result = data['results'][0]['geometry']
            return result['lat'], result['lng']
        else:
            print('No results found for address:', address)
            return None, None
    else:
        print('Error:', response.status_code, response.text)
        return None, None

def geocode_addresses_from_csv(input_csv, output_csv, api_key):
    # Read the CSV file
    df = pd.read_csv(input_csv)
    df= df[500:701]


    # Initialize lists to store the results
    latitudes = []
    longitudes = []

    # Iterate over the addresses and geocode each one
    for address in df['Address']:
        lat, lon = geocode_address(address, api_key)
        latitudes.append(lat)
        longitudes.append(lon)

        # Sleep for a short time to avoid hitting the API rate limit
        time.sleep(.3)


    # Add the results to the dataframe
    df['latitude'] = latitudes
    df['longitude'] = longitudes

    # Write the results to a new CSV file
    df.to_csv(output_csv, index=False)

# Example usage
if __name__ == "__main__":
    input_csv = '/content/LAGOS_crosschecked_address.csv'  # Input CSV file with a column 'address'
    output_csv = 'geocoded_Lagos_addresses_3.csv'  # Output CSV file to save the results
    api_key = '5acb65e575d740ffad2c98a4a6bb5bab'  # Replace with your actual OpenCage API key

    geocode_addresses_from_csv(input_csv, output_csv, api_key)
```
